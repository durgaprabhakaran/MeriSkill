# Day 1: Data Extraction and Cleaning
**Objective:** Import marketing campaign data and clean it for analysis.

```python
import pandas as pd
import numpy as np
```

```python
file_path = r"C:\Users\Durga\Desktop\Career 2025\MeriSkill\Week1\Order_Data_meriskill.csv"
df = pd.read_csv(file_path, low_memory=False)
df.head()
```

```python
# Check data structure
df.info()

# Quick statistics
df.describe()
```

```python
# Check duplicates
print("Duplicate rows:", df.duplicated().sum())

# Check for missing values
df.isnull().sum()
```

```python
# Remove duplicates
df_cleaned = df.drop_duplicates()

# If missing values exist, you can choose:
# Example - filling missing numerical columns with 0 or the mean:
df_cleaned.fillna(0, inplace=True)

# Save cleaned dataset
cleaned_path = r"C:\Users\Durga\Desktop\Career 2025\MeriSkill\Week1\Order_Data_Cleaned.csv"
df_cleaned.to_csv(cleaned_path, index=False)
```

## ✅ Summary:
- Loaded the dataset successfully.
- Checked for duplicates and missing values.
- Removed duplicates and handled missing values.
- Saved cleaned data for analysis as `Order_Data_Cleaned.csv`.
